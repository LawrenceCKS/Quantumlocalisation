```python
filename = "./Data/rss_data.csv"
data = pd.DataFrame( pd.read_csv( filename , header='infer' ) )
data
```

```python
A = data.loc[ (data['Tile']).str.startswith('A1') ]
A.to_csv('A1.csv')
A
```

```python
_view = data.copy()
_view['id'] = _view['SSID'] + "   " + _view['Tile']

_aTable = _view.loc[ (_view['SSID']=="69xXAnonymousXx69-1") ]
_bTable = _view.loc[ (_view['SSID']=="69xXAnonymousXx69-2") ]
_cTable = _view.loc[ (_view['SSID']=="69xXAnonymousXx69-3") ]
_jTable = _view.loc[ (_view['SSID']=="JUb4oa3ahq") ]

fig_size = [10, 13]
```

```python
fig1 , ax1 = plt.subplots( figsize = fig_size )
_count1 = _aTable.groupby(['id'])['id'].count().plot(ax=ax1, kind='barh')
ax1.bar_label(ax1.containers[0])
```

```python
fig2 , ax2 = plt.subplots( figsize = fig_size )
_count2 = _bTable.groupby(['id'])['id'].count().plot(ax=ax2, kind='barh')
ax2.bar_label(ax2.containers[0])
```

```python
fig3 , ax3 = plt.subplots( figsize = fig_size )
_count3 = _cTable.groupby(['id'])['id'].count().plot(ax=ax3, kind='barh')
ax3.bar_label(ax3.containers[0])
```

```python
fig4 , ax4 = plt.subplots( figsize = fig_size )
_count4 = _jTable.groupby(['id'])['id'].count().plot(ax=ax4, kind='barh')
ax4.bar_label(ax4.containers[0])
```
